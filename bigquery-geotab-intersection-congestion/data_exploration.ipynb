Some observations:
1. The same intersections exist in both the training and test sets. There is a lot more in the test set though,  some intersections are there that aren't in the training set (intersection number 3 for instance)
2. While it seems like the points for any given intersection are generally reasonably well spread through the train set...
3. ...observations are sparse enough that you can't get a sufficient handle on an intersection by intersection basis. There are not enough points for most (maybe all) intersection ids to properly distribute between each hour, weekend and weekday and each direction that goes through the intersection
4. I think we need identify bigger blocks of intersections we can evaluate together - neighborhood, intersections along a give street, etc. 
5. We have to be careful about sample biase, however. If you sort by street, for instance, you don't want to be fooled by an intersection that has more early morning instances into thinking that it has less congestion
